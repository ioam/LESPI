## Spatial Profile of LGN Receptive Fields

The spatial profile of LGN Receptive Fields is usually estimated by measuring the response of LGN neurons to moving bar stimuli with varying extents and spatial frequencies. The resulting size tuning curves are then fit using a Difference of Gaussian (DoG) model, expressed as a function of either the spatial frequency (DoGf) or the radius (DoGr).

DoGr: $R = k_c e^{-\frac{r}{r_c}^2} - k_s e^{-\frac{r}{r_s}^2}$

DoGf: $R = k_c e^{-\frac{f}{f_c}^2} - k_s e^{-\frac{f}{f_s}^2}$

More recent papers have also attempted to fit the size response curves with integrated Difference of Gaussian models

iDoG $R(s) = R_0 + K_e \int \int re^{-\frac{r^2}{a}} \, \mathrm{d}r\mathrm{d}\theta - K_i \int\int re^{-\frac{r^2}{b}} \, \mathrm{d}r\mathrm{d}\theta$

where $R_0$ is the spontaneous response rate, $K_e$ the excitatory gain, $K_i$ the inhibitory gain, $a$ the excitatory space constant and $b$ the inhibitory space constant. 

Using this methodology Sceniak et al. 2006 were able to fit excitatory and inhibitory space constants to a sample of 136 neurons in macaque LGN.

The results from Sceniak et al. (2006) represent the best estimates of the spatial properties of LGN receptive fields in macaque.  While excitatory extents vary hugely across various studies the suppressive surround estimates are fairly consistent. Furthermore, the spatial extent of the excitatory CRF centers were found to be contrast invariant, while both the ECRF and CRF suppressive surround extents were found to increase at lower contrast levels. In summary, looking back at all the studies considered here excitatory CRF extents are generally distributed between 0.05-0.5 $\deg$ in radius, while inhibitory CRF and suppressive ECRF radii are distributed distributed anywhere between 0.6-1.5 $\deg$ and the suppression index is quite high (SI>0.8) for 80% cells.

### Feedforward

Using histological and electrophysiological techniques Angelucci et al. (2006) also attempted to quantify the average spatial profile that contribute to the mRF and hsRF of a V1 neuron. This analysis was performed by comparing the so called aggegate receptive field size (ARF) to the hsRF size measured using electrophysiological techniques.

The ARF is defined as:

$ARF_{\deg} = D^{\deg} + RF_{mean}$

where $RF_{mean}$ is the mean RF size of cells recorded at the edge of the injection site, which could reflect the mRF, hsRF or lsRF, and $D^{\deg}$ is defined as:

$D^{\deg} = D_{mm} / MF_{mm/deg} + S_{deg}$

where $D_{mm}$ is the diameter, $MF_{mm/deg}$ is the magnification factor and $S_{deg}$ is the RF scatter at the injection site.

The latest review (Angelucci et al. 2006) has measured the size of the hsRF in V1 neurons of macaques at 2-8 $\deg$ eccentricity as having mean of about 1 $\deg \pm$ 0.1, which is roughly 2.2x larger than the mRF of the same cell based on results from Angelucci et al. (2002) and Levitt et al. (2002). A further analysis by Angelucci et al. (2006b) provides slightly higher estimates with means of 1.09 $\deg$ and 1.41 $\deg$ in layer 4C$\alpha$ and 4A/C$\beta$ respectively.

### Lateral Connectivity

Anatomic data suggests that the spatial spread of lateral connections can be anywhere between 3-10 mm
(on average 6-7 mm) in total length (Angelucci et al., 2002), which is broken down by layer
in table 2.5. Along its principal axis the visuotopic monosynaptic spread of V1 horizontal
connections has a mean of 2.47◦ ± 0.3◦ . This falls well within the range of estimates for the
lsRF as published in a number of studies (Shushruth et al., 2009; Sceniak et al., 1999; Sceniak
and Hawken, 2001), which were fit with the same integrated DoG model and stimulus protocol
as used in the Sceniak et al. (2006) paper on the spatial properties of LGN neurons, reviewed
previously.

In summary, there is strong evidence that lateral connections underly the extent of the lsRF
and mediate a number of effects in the near surround, including contrast dependent size tuning,
colinear facilitation and low contrast suppression. Unlike in other species horizontal connec-
tions are isotropic but do exhibit patchy connectivity in layer 2/3. The extents of horizontal
connections range between 3-10 mm, which averages to around 2.5◦ in visual space.

The GCAL hypercolumn size in this case has simply been centered around the ferret mean, showing the spread in hypercolumn distance. This is because GCAL is not spatially calibrated. In order to actually calibrate the GCAL model another piece of information is needed, the magnification factor at a particular retinotopic location.

### Magnification Factor

The magnification factor describes the visuotopic extent of a degree of visual angle within different brain areas. This conversion changes with eccentricity from the foveal location. In all higher mammals the largest magnification factors are found in foveal location, shrinking significantly as one moves out to the periphery. Most studies record from neurons in para-foveal location, i.e. $2-8 \deg$ out from the fovea. Therefore it makes sense to tune GCAL to a visuotopic location somewhere in this region.
